```python
from os import path, makedirs
import pandas as pd
```

```python
BASE_PATH = path.join("dados", "urbanismo")
CAMINHO_XLSX = path.join(BASE_PATH, "Resumo Painel.xlsx")
CAMINHO_KM = path.join(BASE_PATH, "Quilometragem.xlsx")
```

```python
# Excluir anos 2019 e 2020 e 2021
df_loa = pd.read_excel(CAMINHO_XLSX,
              sheet_name='Loa Atende',
              dtype=str,
              skiprows=1).iloc[:, 1:]

df_loa = df_loa[~df_loa['ano'].isin(['2019', '2020', '2021'])]
df_loa
```

```python
df_atd = pd.read_excel(CAMINHO_XLSX,
              sheet_name='Atendimentos-Viagens',
              dtype=str)
df_atd
```

```python
df_km = pd.read_excel(CAMINHO_KM,
                sheet_name='Sheet1')
df_km
```

```python
output_dir = path.join('data', 'input', 'urbanismo')
if not path.exists(output_dir):
    makedirs(output_dir)

for nome, df in [
    ('loa_atende', df_loa),
    ('atendimentos_viagens', df_atd),
    ('quilometragemKM', df_km)
]:
    filename = path.join(output_dir, nome)
    df.to_csv(f'{filename}.csv', sep=';', decimal=',', index=False)
```
